Выбираемый датасет: House Price Prediction Dataset (https://www.kaggle.com/datasets/zafarali27/house-price-prediction-dataset)

Предметная область: Стоимость домов в зависимости от их характеристик, а именно: площадь дома, количество спален и ванн, количество этажей, год постройки, локация дома, состояние дома, наличие гаража и стоимость.

Источник данных: данные о жилье в Сиднее и Мельбурне.

Характер данных: имитационные.

В датасете представлены след. атрибуты:

1. ID - уникальный числовой тип, который точно дает определять конкретный дом. Здесь перечислено 2000 домов, поэтому и значения будут от 1 до 2000.
2. Площадь - числовой тип, значение указывает площадь дома. Разброс: от 500 до 5000 квадратных футов (1 квадр. фут = 0.09 м^2).
3. Количество спален - числовой тип, значение указывает количество спален в доме. От 1 до 5.
4. Количество ванных комнат - числовой тип, значение указывает количество ванных комнат в доме. От 1 до 4.
5. Количество этажей - числовой тип, значение указывает количество этажей в доме. От 1 до 3.
6. Локация - строковый тип, обозначающий расположение дома, в котором есть четыре варианта записи: Downtown (центр города), Suburban (пригородный), Urban (в городе) и Rural (сельский).
7. Состояние - строковый тип, указывающий состояние дома, в котором есть четыре варианта записи: Excellent (прекрасное), Good (хорошее), Fair (удовлетворительное) и Poor (плохое).
8. Гараж - булевый тип, укуазывает либо Yes - наличие гаража, либо No - отсутствие.
9. Цена - числовой тип данных, который показывает стоимость дома. Принимает значения от 50000 до 1000000$.

Описание задачи анализа: можно спрогназировать цены на жилье в зависимости от множества факторов, начиная с площади жилья, и, заканчивая наличием гаража. 




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных из локального файла (для проверки укажите свой путь)
file_path = r"C:\Users\defau\OneDrive\Desktop\house price prediction dataset.csv"
data = pd.read_csv(file_path)

# Основная информация о данных
print(data.info())

# Преобразование переменных в числовые
data = pd.get_dummies(data)

# Повторная проверка информации о данных после преобразования
print(data.info())

# Анализ каждого атрибута
for column in data.columns:
    print(f'Атрибут: {column}')
    print(f'Среднее значение: {data[column].mean()}')
    print(f'СКО: {data[column].std()}')
 # Проверка на пропущенные значения 
    missing_values = data[column].isnull().sum()
    print(f'Пропущенные значения: {missing_values}')
    
    # Предложение для обработки пропущенных значений 
    if missing_values > 0:
        print(f'Предложение: заменить пропущенные значения на среднее или медиану.')
    
    # Построение гистограммы
    plt.figure(figsize=(10, 4))
    sns.histplot(data[column].dropna(), kde=True)
    plt.title(f'Гистограмма для {column}')
    plt.show()

# Корреляция между параметрами
correlation_matrix = data.corr()
print(correlation_matrix)

# Построение тепловой карты корреляции
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Матрица корреляций')
plt.show()

# Анализ корреляции
threshold = 0.5  # Порог для высокой корреляции
high_corr = [(column1, column2) for column1 in correlation_matrix.columns for column2 in correlation_matrix.columns if abs(correlation_matrix[column1][column2]) > threshold and column1 != column2]
print(f'Высококоррелированные атрибуты: {high_corr}')

# Построение матрицы графиков рассеивания
sns.pairplot(data)
plt.show()
```
